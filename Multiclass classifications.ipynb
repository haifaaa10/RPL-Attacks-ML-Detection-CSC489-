```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score, f1_score, classification_report
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.ensemble import VotingClassifier
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE, ADASYN
from imblearn.under_sampling import RandomUnderSampler
import shap
from sklearn.feature_selection import RFE, mutual_info_classif
import warnings
import os
from datetime import datetime

warnings.filterwarnings('ignore')
plt.style.use('seaborn-v0_8')


# ============================================================
# BASELINE
# ============================================================

class RPLAttackClassifier:
    def __init__(self, data_path):
        self.data_path  = data_path
        self.df = self.X = self.y = None
        self.X_train = self.X_test = self.y_train = self.y_test = None
        self.scaler  = StandardScaler()
        self.models  = {}
        self.results = {}

        self.timestamp   = datetime.now().strftime("%Y%m%d_%H%M%S")
        self.results_dir = f"RPL_Results_{self.timestamp}"
        os.makedirs(self.results_dir, exist_ok=True)
        print(f"Results directory: {self.results_dir}")

    def save_dataframe(self, df, filename):
        path = os.path.join(self.results_dir, f"{filename}.xlsx")
        df.to_excel(path, index=False)
        print(f"Saved: {path}")

    def save_plot(self, filename):
        path = os.path.join(self.results_dir, f"{filename}.pdf")
        plt.savefig(path, dpi=300, bbox_inches='tight')
        print(f"Saved: {path}")

    def load_and_preprocess_data(self):
        print("Loading dataset...")
        self.df = pd.read_csv(self.data_path)

        imputer = SimpleImputer(strategy='median')
        if 'Psent' in self.df.columns:
            self.df['Psent'] = imputer.fit_transform(self.df[['Psent']])

        self.df['Network_encoded'] = LabelEncoder().fit_transform(self.df['Network'])
        self.df['OF_encoded']      = LabelEncoder().fit_transform(self.df['OF'])

        feature_cols = ['IDs', 'DIOs', 'DISs', 'DAOs', 'Precv', 'Pforw',
                        'Pdrop', 'Psent', 'Network_encoded', 'OF_encoded']
        feature_cols = [c for c in feature_cols if c in self.df.columns]

        self.X = self.df[feature_cols]
        self.y = self.df['Attack']

        print(f"Dataset shape: {self.df.shape}")
        print(f"Class distribution:\n{self.y.value_counts().sort_index()}")
        return self.X, self.y

    def split_and_scale_data(self):
        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(
            self.X, self.y, test_size=0.2, random_state=42, stratify=self.y
        )
        self.X_train_scaled = self.scaler.fit_transform(self.X_train)
        self.X_test_scaled  = self.scaler.transform(self.X_test)
        print(f"Train: {self.X_train.shape}, Test: {self.X_test.shape}")

    def initialize_models(self):
        rf  = RandomForestClassifier(random_state=42, n_estimators=100)
        gb  = GradientBoostingClassifier(random_state=42)
        xgb = XGBClassifier(random_state=42, eval_metric='mlogloss')

        self.models = {
            'Random Forest':          rf,
            'XGBoost':                xgb,
            'Voting Classifier (Soft)': VotingClassifier(
                estimators=[('rf', rf), ('xgb', xgb), ('gb', gb)],
                voting='soft'
            )
        }
        print(f"Models initialized: {list(self.models.keys())}")

    def train_and_evaluate(self):
        print("\n" + "="*80)
        print("BASELINE RESULTS")
        print("="*80)

        rows = []
        for name, model in self.models.items():
            model.fit(self.X_train, self.y_train)
            y_pred      = model.predict(self.X_test)
            y_train_pred = model.predict(self.X_train)

            acc          = accuracy_score(self.y_test, y_pred)
            train_acc    = accuracy_score(self.y_train, y_train_pred)
            weighted_f1  = f1_score(self.y_test, y_pred, average='weighted', zero_division=0)
            gap          = train_acc - acc

            cv_folds = 3 if 'Voting' in name else 5
            cv       = cross_val_score(model, self.X_train, self.y_train,
                                       cv=cv_folds, scoring='accuracy')

            self.results[name] = {
                'model': model, 'accuracy': acc, 'train_accuracy': train_acc,
                'weighted_f1': weighted_f1, 'overfitting_gap': gap,
                'cv_mean': cv.mean(), 'cv_std': cv.std(),
                'model_type': 'Ensemble' if 'Voting' in name else 'Individual'
            }

            rows.append({
                'Model': name, 'Type': self.results[name]['model_type'],
                'Test Accuracy': round(acc, 4),
                'Weighted F1':   round(weighted_f1, 4),
                'Overfitting Gap': round(gap, 4),
                'CV Accuracy':   round(cv.mean(), 4),
                'CV Std':        round(cv.std(), 5),
            })

            print(f"{name}: Acc={acc:.4f}, WF1={weighted_f1:.4f}, "
                  f"Gap={gap:.4f}, CV={cv.mean():.4f}±{cv.std():.5f}")

        df = pd.DataFrame(rows).sort_values('Test Accuracy', ascending=False)
        print("\nBASELINE TABLE (matches original paper format):")
        print(df.to_string(index=False))
        self.save_dataframe(df, "Table1_Baseline_Results")
        return df


# ============================================================
# PHASE 1: CLASS IMBALANCE STUDY
# ============================================================

class ImbalanceStudy:
    def __init__(self, classifier):
        self.clf         = classifier
        self.X           = classifier.X
        self.y           = classifier.y
        self.results_dir = classifier.results_dir
        self.datasets    = {}   # scenario → (X, y)
        self.results     = {}   # scenario → technique → model → metrics

    def save_dataframe(self, df, filename):
        path = os.path.join(self.results_dir, f"{filename}.xlsx")
        df.to_excel(path, index=False)
        print(f"Saved: {path}")

    def save_plot(self, filename):
        path = os.path.join(self.results_dir, f"{filename}.pdf")
        plt.savefig(path, dpi=300, bbox_inches='tight')
        print(f"Saved: {path}")

    def create_scenarios(self):
        """
        Balanced uses original data + original split (matches baseline exactly).
        Imbalanced scenarios are constructed by adjusting class counts.
        """
        np.random.seed(42)
        print("\n" + "="*80)
        print("PHASE 1: CREATING IMBALANCED SCENARIOS")
        print("="*80)

        # Balanced: original data, no resampling
        self.datasets['Balanced'] = (self.X, self.y)

        imbalanced = {
            'Mild (70/30)':     {0: 21168, 1: 2268, 2: 2268, 3: 2268, 4: 2268},
            'Moderate (85/15)': {0: 25704, 1: 1134, 2: 1134, 3: 1134, 4: 1134},
            'Severe (95/5)':    {0: 28728, 1: 378,  2: 378,  3: 378,  4: 378},
        }

        for name, counts in imbalanced.items():
            X_list, y_list = [], []
            for cls, n in counts.items():
                X_c = self.X[self.y == cls]
                y_c = self.y[self.y == cls]
                idx = np.random.choice(len(X_c), size=n, replace=(n > len(X_c)))
                X_list.append(X_c.iloc[idx])
                y_list.append(y_c.iloc[idx])
            Xs = pd.concat(X_list, ignore_index=True)
            ys = pd.concat(y_list, ignore_index=True)
            self.datasets[name] = (Xs, ys)
            print(f"{name}: total={len(ys)}, "
                  f"benign={counts[0]}, attack/class={counts[1]}")

    def run(self):
        print("\n" + "="*80)
        print("PHASE 1: EVALUATING RESAMPLING TECHNIQUES")
        print("="*80)

        techniques = {
            'No Resampling':      None,
            'SMOTE':              SMOTE(random_state=42),
            'ADASYN':             ADASYN(random_state=42),
            'Random Undersample': RandomUnderSampler(random_state=42),
            'Gaussian Noise':     'gaussian',
        }

        def gaussian_oversample(X, y, std=0.05):
            classes, counts = np.unique(y, return_counts=True)
            max_n = counts.max()
            Xl, yl = [X], [y]
            for cls, cnt in zip(classes, counts):
                if cnt < max_n:
                    Xc = X[y == cls]
                    idx = np.random.choice(len(Xc), max_n - cnt, replace=True)
                    Xl.append(Xc[idx] + np.random.normal(0, std, Xc[idx].shape))
                    yl.append(np.full(max_n - cnt, cls))
            return np.vstack(Xl), np.concatenate(yl)

        scaler = StandardScaler()

        for scenario, (X_sc, y_sc) in self.datasets.items():
            self.results[scenario] = {}
            is_balanced = (scenario == 'Balanced')

            if is_balanced:
                # Use original train/test split — matches baseline exactly
                X_train_base = self.clf.X_train_scaled
                X_test_base  = self.clf.X_test_scaled
                y_train_base = self.clf.y_train
                y_test_base  = self.clf.y_test
            else:
                X_scaled = scaler.fit_transform(X_sc)

            for tech_name, resampler in techniques.items():
                if is_balanced:
                    # For balanced scenario all techniques use the original split
                    X_tr, X_te, y_tr, y_te = (
                        X_train_base, X_test_base,
                        y_train_base, y_test_base
                    )
                    model = XGBClassifier(random_state=42, eval_metric='mlogloss')

                else:
                    if resampler is None:
                        # No resampling — just split
                        X_tr, X_te, y_tr, y_te = train_test_split(
                            X_scaled, y_sc, test_size=0.2,
                            random_state=42, stratify=y_sc
                        )
                        model = XGBClassifier(random_state=42, eval_metric='mlogloss')

                    elif resampler == 'gaussian':
                        # Gaussian noise oversampling
                        np.random.seed(42)
                        X_res, y_res = gaussian_oversample(X_scaled, y_sc.values)
                        X_tr, X_te, y_tr, y_te = train_test_split(
                            X_res, y_res, test_size=0.2,
                            random_state=42, stratify=y_res
                        )
                        model = XGBClassifier(random_state=42, eval_metric='mlogloss')

                    else:
                        # SMOTE, ADASYN, Random Undersample
                        try:
                            X_res, y_res = resampler.fit_resample(X_scaled, y_sc)
                        except Exception as e:
                            print(f"  {tech_name} failed for {scenario}: {e}")
                            continue
                        X_tr, X_te, y_tr, y_te = train_test_split(
                            X_res, y_res, test_size=0.2,
                            random_state=42, stratify=y_res
                        )
                        model = XGBClassifier(random_state=42, eval_metric='mlogloss')

                try:
                    model.fit(X_tr, y_tr)
                    y_pred    = model.predict(X_te)
                    y_tr_pred = model.predict(X_tr)

                    acc         = accuracy_score(y_te, y_pred)
                    train_acc   = accuracy_score(y_tr, y_tr_pred)
                    gap         = train_acc - acc
                    macro_f1    = f1_score(y_te, y_pred, average='macro',    zero_division=0)
                    weighted_f1 = f1_score(y_te, y_pred, average='weighted', zero_division=0)
                    per_class   = f1_score(y_te, y_pred, average=None,       zero_division=0)

                    self.results[scenario][tech_name] = {
                        'accuracy':        acc,
                        'train_accuracy':  train_acc,
                        'overfitting_gap': gap,
                        'macro_f1':        macro_f1,
                        'weighted_f1':     weighted_f1,
                        'per_class_f1':    per_class,
                    }
                    print(f"  [{scenario}] {tech_name}: "
                          f"Acc={acc:.4f}, Gap={gap:.4f}, "
                          f"MacroF1={macro_f1:.4f}, WeightedF1={weighted_f1:.4f}")

                except Exception as e:
                    print(f"  [{scenario}] {tech_name} model failed: {e}")

    def save_and_plot(self):
        self._save_summary_table()
        self._plot_macro_f1_heatmap()
        self._plot_accuracy_vs_macrof1()
        self._plot_per_class_f1_severe()

    def _save_summary_table(self):
        """
        Table 2: Main Phase 1 results — same format as original paper Table 5
        Columns: Technique, Scenario, Accuracy, Weighted F1, Macro F1,
                 Overfitting Gap, CV Accuracy, CV Std
        Note: CV is only computed for balanced.
              For imbalanced scenarios, CV is omitted (different data splits).
        """
        rows = []
        for scenario, techniques in self.results.items():
            for tech, metrics in techniques.items():
                rows.append({
                    'Scenario':        scenario,
                    'Technique':       tech,
                    'Accuracy':        round(metrics['accuracy'],        4),
                    'Weighted F1':     round(metrics['weighted_f1'],     4),
                    'Macro F1':        round(metrics['macro_f1'],        4),
                    'Overfitting Gap': round(metrics['overfitting_gap'], 4),
                })
        df = pd.DataFrame(rows)
        print("\n" + "="*100)
        print("PHASE 1 SUMMARY TABLE")
        print("="*100)
        print(df.to_string(index=False))
        self.save_dataframe(df, "Table2_Phase1_Imbalance_Results")

    def _plot_macro_f1_heatmap(self):
        scenarios  = list(self.results.keys())
        techniques = list(self.results[scenarios[0]].keys())
        data = []
        for tech in techniques:
            row = []
            for sc in scenarios:
                try:
                    row.append(self.results[sc][tech]['macro_f1'])
                except KeyError:
                    row.append(np.nan)
            data.append(row)

        df_h = pd.DataFrame(data, index=techniques, columns=scenarios)

        plt.figure(figsize=(12, 5))
        sns.heatmap(df_h, annot=True, fmt='.4f', cmap='RdYlGn',
                    vmin=0.5, vmax=1.0, linewidths=0.5, annot_kws={"size": 11})
        plt.title('Figure 1: Phase 1 — Macro-F1 Score\n'
                  'Resampling Technique vs Imbalance Scenario (XGBoost)')
        plt.xlabel('Imbalance Scenario')
        plt.ylabel('Resampling Technique')
        plt.tight_layout()
        self.save_plot("Figure1_Phase1_MacroF1_Heatmap")
        plt.show()

    def _plot_accuracy_vs_macrof1(self):
        scenarios  = list(self.results.keys())
        techniques = list(self.results[scenarios[0]].keys())

        fig, axes = plt.subplots(1, 2, figsize=(14, 5))

        for tech in techniques:
            accs, f1s = [], []
            for sc in scenarios:
                try:
                    accs.append(self.results[sc][tech]['accuracy'])
                    f1s.append( self.results[sc][tech]['macro_f1'])
                except KeyError:
                    accs.append(np.nan)
                    f1s.append(np.nan)
            axes[0].plot(scenarios, accs, marker='o', linewidth=2, label=tech)
            axes[1].plot(scenarios, f1s,  marker='o', linewidth=2, label=tech)

        for ax, ylabel in zip(axes, ['Accuracy', 'Macro F1']):
            ax.set_xlabel('Imbalance Scenario')
            ax.set_ylabel(ylabel)
            ax.set_ylim(0.85, 1.02)
            ax.grid(True, alpha=0.3)
            ax.legend(fontsize=8)
            ax.tick_params(axis='x', rotation=10)

        axes[0].set_title('Accuracy (misleadingly stable)')
        axes[1].set_title('Macro F1 (reveals true degradation)')
        plt.suptitle('Figure 2: Accuracy vs Macro-F1 Under Increasing Imbalance (XGBoost)',
                     fontsize=12)
        plt.tight_layout()
        self.save_plot("Figure2_Accuracy_vs_MacroF1")
        plt.show()

    def _plot_per_class_f1_severe(self):
        scenario    = 'Severe (95/5)'
        class_names = ['Normal', 'Blackhole', 'Decr. Rank', 'DODAG Ver.', 'Flooding']
        techniques  = list(self.results.get(scenario, {}).keys())
        x     = np.arange(len(class_names))
        width = 0.15

        plt.figure(figsize=(13, 5))
        for i, tech in enumerate(techniques):
            try:
                pc = self.results[scenario][tech]['per_class_f1']
                while len(pc) < 5:
                    pc = np.append(pc, 0)
                plt.bar(x + i * width, pc[:5], width, label=tech, alpha=0.85)
            except KeyError:
                continue

        plt.xlabel('Attack Class')
        plt.ylabel('F1 Score')
        plt.title('Figure 3: Per-Class F1 Under Severe Imbalance (95% Benign) — XGBoost')
        plt.xticks(x + width * len(techniques) / 2, class_names)
        plt.legend(loc='lower right', fontsize=8)
        plt.ylim(0, 1.12)
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        self.save_plot("Figure3_PerClass_F1_Severe")
        plt.show()


# ============================================================
# PHASE 2: FEATURE SELECTION STUDY
# ============================================================

class FeatureSelectionStudy:
    def __init__(self, classifier):
        self.clf          = classifier
        self.X_train      = classifier.X_train
        self.X_test       = classifier.X_test
        self.y_train      = classifier.y_train
        self.y_test       = classifier.y_test
        self.feature_names = list(classifier.X.columns)
        self.results_dir  = classifier.results_dir

        self.shap_ranking = None
        self.rfe_ranking  = None
        self.mi_ranking   = None
        self.subset_results = {}

    def run_all_rankings(self):
        self._run_shap()
        self._run_rfe()
        self._run_mi()
        self._plot_ranking_comparison()

    def _run_shap(self):
        print("\n" + "="*80)
        print("PHASE 2: SHAP FEATURE RANKING")
        print("="*80)
        rf = self.clf.models['Random Forest']
        X_vals = self.X_test.values[:500]
        explainer = shap.TreeExplainer(rf)
        sv = explainer.shap_values(X_vals)

        if isinstance(sv, list):
            mean_abs = np.mean([np.abs(s).mean(axis=0) for s in sv], axis=0)
        elif len(sv.shape) == 3:
            mean_abs = np.abs(sv).mean(axis=(0, 2))
        else:
            mean_abs = np.abs(sv).mean(axis=0)

        df = pd.DataFrame({'Feature': self.feature_names, 'SHAP': mean_abs}
                          ).sort_values('SHAP', ascending=False).reset_index(drop=True)
        print(df.to_string())
        self.shap_ranking     = df['Feature'].tolist()
        self.shap_importance_df = df

    def _run_rfe(self):
        print("\n" + "="*80)
        print("PHASE 2: RFE FEATURE RANKING")
        print("="*80)
        rfe = RFE(RandomForestClassifier(n_estimators=50, random_state=42),
                  n_features_to_select=1, step=1)
        rfe.fit(self.X_train, self.y_train)
        df = pd.DataFrame({'Feature': self.feature_names, 'RFE_Rank': rfe.ranking_}
                          ).sort_values('RFE_Rank').reset_index(drop=True)
        print(df.to_string())
        self.rfe_ranking     = df['Feature'].tolist()
        self.rfe_importance_df = df

    def _run_mi(self):
        print("\n" + "="*80)
        print("PHASE 2: MUTUAL INFORMATION FEATURE RANKING")
        print("="*80)
        mi = mutual_info_classif(self.X_train, self.y_train, random_state=42)
        df = pd.DataFrame({'Feature': self.feature_names, 'MI': mi}
                          ).sort_values('MI', ascending=False).reset_index(drop=True)
        print(df.to_string())
        self.mi_ranking     = df['Feature'].tolist()
        self.mi_importance_df = df

    def _plot_ranking_comparison(self):
        fig, axes = plt.subplots(1, 3, figsize=(16, 5))

        axes[0].barh(self.shap_importance_df['Feature'][::-1],
                     self.shap_importance_df['SHAP'][::-1],
                     color='steelblue', alpha=0.85)
        axes[0].set_title('SHAP')
        axes[0].set_xlabel('Mean |SHAP Value|')
        axes[0].grid(True, alpha=0.3)

        axes[1].barh(self.rfe_importance_df['Feature'][::-1],
                     self.rfe_importance_df['RFE_Rank'][::-1],
                     color='seagreen', alpha=0.85)
        axes[1].set_title('RFE (lower rank = more important)')
        axes[1].set_xlabel('Rank')
        axes[1].grid(True, alpha=0.3)

        axes[2].barh(self.mi_importance_df['Feature'][::-1],
                     self.mi_importance_df['MI'][::-1],
                     color='darkorange', alpha=0.85)
        axes[2].set_title('Mutual Information')
        axes[2].set_xlabel('MI Score')
        axes[2].grid(True, alpha=0.3)

        plt.suptitle('Figure 4: Feature Ranking Comparison — SHAP, RFE, Mutual Information',
                     fontsize=12)
        plt.tight_layout()
        path = os.path.join(self.results_dir, "Figure4_Feature_Ranking_Comparison.pdf")
        plt.savefig(path, dpi=300, bbox_inches='tight')
        print(f"Saved: {path}")
        plt.show()

    def test_subsets(self):
        """
        Tests feature subsets of size 3, 5, 7, 10.
        Reports SAME metrics as original paper Table 5:
        Weighted F1, Overfitting Gap, CV Accuracy, CV Std.
        Also reports Accuracy and Macro F1 for completeness.
        Uses original train/test split on balanced data.
        """
        print("\n" + "="*80)
        print("PHASE 2: FEATURE SUBSET EVALUATION")
        print("="*80)

        subset_sizes = [3, 5, 7, 10]
        methods = {
            'SHAP':       self.shap_ranking,
            'RFE':        self.rfe_ranking,
            'MutualInfo': self.mi_ranking,
        }
        model_names = ['Voting Classifier (Soft)', 'XGBoost', 'Random Forest']
        scaler = StandardScaler()

        for method_name, ranking in methods.items():
            self.subset_results[method_name] = {}
            print(f"\n  Method: {method_name}")

            for n in subset_sizes:
                selected = ranking[:n]

                if n == 10:
                   X_tr = self.X_train.values
                   X_te = self.X_test.values
                else:
                   X_tr = scaler.fit_transform(self.X_train[selected])
                   X_te = scaler.transform(self.X_test[selected])

                self.subset_results[method_name][n] = {}

                for mname in model_names:
                    model = self._fresh_model(mname)
                    try:
                        model.fit(X_tr, self.y_train)
                        y_pred    = model.predict(X_te)
                        y_tr_pred = model.predict(X_tr)

                        acc         = accuracy_score(self.y_test, y_pred)
                        train_acc   = accuracy_score(self.y_train, y_tr_pred)
                        gap         = train_acc - acc
                        weighted_f1 = f1_score(self.y_test, y_pred,
                                               average='weighted', zero_division=0)
                        macro_f1    = f1_score(self.y_test, y_pred,
                                               average='macro', zero_division=0)

                        cv_folds = 3 if 'Voting' in mname else 5
                        cv = cross_val_score(model, X_tr, self.y_train,
                                             cv=cv_folds, scoring='accuracy')

                        self.subset_results[method_name][n][mname] = {
                            'accuracy': acc, 'train_accuracy': train_acc,
                            'overfitting_gap': gap,
                            'weighted_f1': weighted_f1, 'macro_f1': macro_f1,
                            'cv_mean': cv.mean(), 'cv_std': cv.std(),
                        }

                        print(f"    n={n}, {mname}: "
                              f"Acc={acc:.4f}, WF1={weighted_f1:.4f}, "
                              f"Gap={gap:.4f}, CV={cv.mean():.4f}±{cv.std():.5f}")

                    except Exception as e:
                        print(f"    n={n}, {mname} failed: {e}")

        self._save_subset_table()
        self._plot_subset_curves()

    def _fresh_model(self, name):
        if 'Voting' in name:
            return VotingClassifier(
                estimators=[
                    ('rf',  RandomForestClassifier(random_state=42, n_estimators=100)),
                    ('xgb', XGBClassifier(random_state=42, eval_metric='mlogloss')),
                    ('gb',  GradientBoostingClassifier(random_state=42))
                ], voting='soft'
            )
        elif 'XGBoost' in name:
            return XGBClassifier(random_state=42, eval_metric='mlogloss')
        else:
            return RandomForestClassifier(random_state=42, n_estimators=100)

    def _save_subset_table(self):
        """
        Table 5 (matching original paper format) + Macro F1 column added.
        Overhead reduction = (10 - n) / 10 * 100 %
        """
        rows = []
        for method, sizes in self.subset_results.items():
            for n, models in sizes.items():
                for mname, m in models.items():
                    rows.append({
                        'Selection Method': method,
                        'N Features':       n,
                        'Model':            mname,
                        'Test Accuracy':    round(m['accuracy'],        4),
                        'Weighted F1':      round(m['weighted_f1'],     4),
                        'Macro F1':         round(m['macro_f1'],        4),
                        'Overfitting Gap':  round(m['overfitting_gap'], 4),
                        'CV Accuracy':      round(m['cv_mean'],         4),
                        'CV Std':           round(m['cv_std'],          5),
                        'Overhead Reduction (%)': f"{int((10 - n) / 10 * 100)}%",
                    })

        df = pd.DataFrame(rows)
        print("\n" + "="*120)
        print("PHASE 2 SUMMARY TABLE — SHAP method (same format as original paper Table 5)")
        print("="*120)
        shap_df = df[df['Selection Method'] == 'SHAP'].drop(
            columns=['Selection Method']).reset_index(drop=True)
        print(shap_df.to_string(index=False))
        self.clf.save_dataframe(df, "Table3_Phase2_Feature_Subset_Results")

    def _plot_subset_curves(self):
        subset_sizes = [3, 5, 7, 10]
        model_names  = ['Voting Classifier (Soft)', 'XGBoost', 'Random Forest']
        methods      = list(self.subset_results.keys())

        fig, axes = plt.subplots(1, len(methods), figsize=(16, 5), sharey=True)

        for ax, method in zip(axes, methods):
            for mname in model_names:
                accs = []
                for n in subset_sizes:
                    try:
                        accs.append(self.subset_results[method][n][mname]['weighted_f1'])
                    except KeyError:
                        accs.append(np.nan)
                ax.plot(subset_sizes, accs, marker='o', linewidth=2, label=mname)

            ax.axhline(y=0.9944, color='gray', linestyle='--',
                       alpha=0.6, label='RF baseline (10 feat)')
            ax.set_title(f'{method}')
            ax.set_xlabel('Number of Features')
            ax.set_ylabel('Weighted F1')
            ax.set_xticks(subset_sizes)
            ax.set_ylim(0.88, 1.01)
            ax.grid(True, alpha=0.3)
            ax.legend(fontsize=8)

        plt.suptitle('Figure 5: Weighted F1 vs Feature Subset Size\n'
                     'Dashed line = Random Forest full-feature baseline (0.9944)',
                     fontsize=12)
        plt.tight_layout()
        path = os.path.join(self.clf.results_dir, "Figure5_WeightedF1_vs_Features.pdf")
        plt.savefig(path, dpi=300, bbox_inches='tight')
        print(f"Saved: {path}")
        plt.show()

    def combined_key_figure(self, imbalance_study):
        """
        Key figure: Macro F1 across feature subset size × imbalance scenario.
        Balanced uses original split. Imbalanced uses Gaussian noise oversampling.
        """
        print("\n" + "="*80)
        print("PHASE 1+2 COMBINED: KEY FIGURE")
        print("="*80)

        def gaussian_oversample(X, y, std=0.05):
            classes, counts = np.unique(y, return_counts=True)
            max_n = counts.max()
            Xl, yl = [X], [y]
            for cls, cnt in zip(classes, counts):
                if cnt < max_n:
                    Xc  = X[y == cls]
                    idx = np.random.choice(len(Xc), max_n - cnt, replace=True)
                    Xl.append(Xc[idx] + np.random.normal(0, std, Xc[idx].shape))
                    yl.append(np.full(max_n - cnt, cls))
            return np.vstack(Xl), np.concatenate(yl)

        subset_sizes = [3, 5, 7, 10]
        scenarios    = list(imbalance_study.datasets.keys())
        scaler       = StandardScaler()
        ranking      = self.shap_ranking
        data         = np.zeros((len(subset_sizes), len(scenarios)))

        for i, n in enumerate(subset_sizes):
            feat_idx = [self.clf.X.columns.tolist().index(f) for f in ranking[:n]]

            for j, scenario in enumerate(scenarios):
                X_sc, y_sc = imbalance_study.datasets[scenario]

                if scenario == 'Balanced':
                     if n == 10:
                         # No scaling — matches baseline exactly
                         X_tr = self.clf.X_train.iloc[:, feat_idx].values
                         X_te = self.clf.X_test.iloc[:, feat_idx].values
                     else:
                         X_tr = scaler.fit_transform(
                             self.clf.X_train.iloc[:, feat_idx].values)
                         X_te = scaler.transform(
                             self.clf.X_test.iloc[:, feat_idx].values)
                     y_tr = self.clf.y_train.values
                     y_te = self.clf.y_test.values
                else:
                    X_sub    = X_sc.iloc[:, feat_idx].values
                    X_scaled = scaler.fit_transform(X_sub)
                    np.random.seed(42)
                    X_res, y_res = gaussian_oversample(X_scaled, y_sc.values)
                    X_tr, X_te, y_tr, y_te = train_test_split(
                        X_res, y_res, test_size=0.2,
                        random_state=42, stratify=y_res
                    )

                model = XGBClassifier(random_state=42, eval_metric='mlogloss')
                model.fit(X_tr, y_tr)
                y_pred = model.predict(X_te)
                f1 = f1_score(y_te, y_pred, average='macro', zero_division=0)
                data[i, j] = f1
                print(f"  n={n}, {scenario}: Macro F1 = {f1:.4f}")

        df_heatmap = pd.DataFrame(
            data,
            index=[f'Top {n} features' for n in subset_sizes],
            columns=scenarios
        )

        print("\nCOMBINED HEATMAP DATA:")
        print(df_heatmap.round(4).to_string())
        self.clf.save_dataframe(
            df_heatmap.reset_index().rename(columns={'index': 'Feature Subset'}),
            "Table4_Combined_Heatmap"
        )

        plt.figure(figsize=(11, 5))
        sns.heatmap(df_heatmap, annot=True, fmt='.4f', cmap='RdYlGn',
                    vmin=0.90, vmax=1.0, linewidths=0.5, annot_kws={"size": 13})
        plt.title(
            'Figure 6: Macro F1 — Feature Subset Size × Imbalance Scenario\n'
            'Model: XGBoost | Feature Selection: SHAP | '
            'Resampling: Gaussian Noise (imbalanced) / Original split (balanced)',
            fontsize=11
        )
        plt.xlabel('Imbalance Scenario')
        plt.ylabel('Feature Subset')
        plt.tight_layout()
        path = os.path.join(self.clf.results_dir, "Figure6_Combined_Key_Figure.pdf")
        plt.savefig(path, dpi=300, bbox_inches='tight')
        print(f"Saved: {path}")
        plt.show()

        return df_heatmap


# ============================================================
# MAIN
# ============================================================

def main():
    print("="*80)
    print("RPL ATTACK DETECTION EXTENSION")
    print("Phase 1: Class Imbalance | Phase 2: Feature Selection")
    print("="*80)

    data_path  = "rpl_multiclass_attack_dataset_30240.csv"
    clf = RPLAttackClassifier(data_path)
    clf.load_and_preprocess_data()
    clf.split_and_scale_data()
    clf.initialize_models()
    clf.train_and_evaluate()

    print("\n" + "="*80)
    print("STARTING EXTENSIONS")
    print("="*80)

    # Phase 1
    imb = ImbalanceStudy(clf)
    imb.create_scenarios()
    imb.run()
    imb.save_and_plot()

    # Phase 2
    feat = FeatureSelectionStudy(clf)
    feat.run_all_rankings()
    feat.test_subsets()

    # Combined
    feat.combined_key_figure(imb)

    print("\n" + "="*80)
    print(f"ALL DONE — results saved in: {clf.results_dir}")
    print("="*80)
    print("\nOutputs produced:")
    print("  Table1_Baseline_Results.xlsx")
    print("  Table2_Phase1_Imbalance_Results.xlsx")
    print("  Table3_Phase2_Feature_Subset_Results.xlsx")
    print("  Table4_Combined_Heatmap.xlsx")
    print("  Figure1_Phase1_MacroF1_Heatmap.pdf")
    print("  Figure2_Accuracy_vs_MacroF1.pdf")
    print("  Figure3_PerClass_F1_Severe.pdf")
    print("  Figure4_Feature_Ranking_Comparison.pdf")
    print("  Figure5_WeightedF1_vs_Features.pdf")
    print("  Figure6_Combined_Key_Figure.pdf")


if __name__ == "__main__":
    main()
```
