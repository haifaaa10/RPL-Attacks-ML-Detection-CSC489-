```python
"""
=============================================================================
RPL ATTACK DETECTION — IoT-RPL 2021 DATASET (COMPLETE FINAL)
Phase 1: Class Imbalance Study | Phase 2: Feature Selection Study
=============================================================================
Dataset : IoT-RPL 2021 (Dhifallah et al., Mendeley Data, 2024,
          DOI 10.17632/4rcbbry2sc.1)
          Engineered by iot_rpl_feature_engineering.py

Sampling strategy:
  Training : 100K stratified sample per seed
  Test     : 50K  stratified sample (fixed, seed=0)

Feedback issues addressed:
  Issue 1  — IoT-RPL 2021 dataset used
  Issue 3  — Training time, inference time, model size measured
  Issue 4  — 10 seeds, 95% bootstrap CI, Wilcoxon signed-rank tests
  Issue 5  — Class Weighted added as 6th technique
  Issue 6  — Sensitivity sweep {0.05,0.1,0.2,0.5}
  Issue 7  — Spearman rho + Kendall tau between SHAP, RFE, MI rankings
  Issue 8  — Per-stratum results by attack class

Features (13 engineered rolling-window columns):
  roll_total, roll_n_DIS, roll_n_DIO, roll_n_DAO,
  roll_ratio_DIS, roll_ratio_DIO, roll_ratio_DAO,
  roll_unique_to, roll_avg_doagid2, roll_avg_doagid3,
  roll_avg_dio, roll_avg_obj, roll_avg_life

Classes (5): Normal, Flooding, Blackhole, Version, Rank

Imbalance scenarios (Normal-dominant):
  Natural          = actual IoT-RPL 2021 distribution
  Mild (70/30)     = Normal 70%, each attack ~7.5%
  Moderate (85/15) = Normal 85%, each attack ~3.75%
  Severe (95/5)    = Normal 95%, each attack ~1.25%

Resampling techniques (6):
  No Resampling, SMOTE (k=5), ADASYN (n=5), Random Undersample,
  Gaussian Noise (sigma=0.05), Class Weighted

Feature subset sizes: 3, 5, 7, 10, 13

Outputs:
  Table1_Baseline_Results.xlsx
  Table2_Phase1_Imbalance_Results.xlsx
  Table2b_Wilcoxon_Tests.xlsx
  Table2c_Sigma_Sensitivity.xlsx
  Table3_Phase2_Feature_Subset_Results.xlsx
  Table3b_Efficiency_Metrics.xlsx
  Table3c_Per_Stratum_Results.xlsx
  Table4_Combined_Heatmap.xlsx
  Table5_Ranking_Correlations.xlsx
  Figure1_Phase1_MacroF1_Heatmap.pdf
  Figure2_MacroF1_With_CI.pdf
  Figure3_PerClass_F1_Severe.pdf
  Figure4_Feature_Ranking_Comparison.pdf
  Figure5_WeightedF1_vs_Features.pdf
  Figure6_Combined_Key_Figure.pdf
  Figure7_Per_Seed_Scatter.pdf
  Figure8_Inference_Time_vs_Features.pdf
  Figure_Per_Stratum.pdf
=============================================================================
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import tracemalloc
from scipy.stats import spearmanr, kendalltau, wilcoxon
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, f1_score
from sklearn.ensemble import (RandomForestClassifier,
                               GradientBoostingClassifier,
                               VotingClassifier)
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE, ADASYN
from imblearn.under_sampling import RandomUnderSampler
import shap
from sklearn.feature_selection import RFE, mutual_info_classif
import warnings
import os
import time
from datetime import datetime

warnings.filterwarnings('ignore')
plt.style.use('seaborn-v0_8')

# ============================================================
# CONFIGURATION  — only section that needs editing
# ============================================================

TRAIN_PATH     = "iot_rpl_engineered_train.csv"
TEST_PATH      = "iot_rpl_engineered_test.csv"
LABEL_COL      = "label"
TRAIN_SAMPLE   = 100_000
TEST_SAMPLE    = 50_000
SEEDS = list(range(42, 52))
N_BOOT         = 10_000
GAUSSIAN_SIGMA = 0.05          # Issue 6: documented; swept in sensitivity analysis

FEATURE_COLS   = [
    'roll_total', 'roll_n_DIS', 'roll_n_DIO', 'roll_n_DAO',
    'roll_ratio_DIS', 'roll_ratio_DIO', 'roll_ratio_DAO',
    'roll_unique_to', 'roll_avg_doagid2', 'roll_avg_doagid3',
    'roll_avg_dio', 'roll_avg_obj', 'roll_avg_life'
]
SUBSET_SIZES   = [3, 5, 7, 10, 13]
SIGMA_SWEEP    = [0.05, 0.1, 0.2, 0.5]   # Issue 6 sensitivity sweep


# ============================================================
# UTILITY FUNCTIONS
# ============================================================

def bootstrap_ci(scores, n_boot=N_BOOT, ci=95):
    scores = np.array(scores)
    if len(scores) == 1:
        return scores[0], scores[0], scores[0]
    boot  = np.random.choice(scores, size=(n_boot, len(scores)), replace=True)
    means = boot.mean(axis=1)
    lo    = np.percentile(means, (100 - ci) / 2)
    hi    = np.percentile(means, 100 - (100 - ci) / 2)
    return scores.mean(), lo, hi


def ci_str(scores):
    m, lo, hi = bootstrap_ci(scores)
    return f"{m:.4f} [{lo:.4f}, {hi:.4f}]"


def wilcoxon_test(a, b):
    try:
        _, p = wilcoxon(a, b)
        return p
    except Exception:
        return np.nan


def stratified_sample(X, y, n, seed):
    sss = StratifiedShuffleSplit(n_splits=1, train_size=n, random_state=seed)
    idx, _ = next(sss.split(X, y))
    return X[idx], y[idx]


def measure_inference_time(model, X, n_repeats=10):
    """Inference time in microseconds per sample."""
    times = []
    for _ in range(n_repeats):
        t0 = time.perf_counter()
        model.predict(X)
        times.append(time.perf_counter() - t0)
    return (np.mean(times) / len(X)) * 1e6


def measure_model_size(model):
    """Model size in KB."""
    return len(pickle.dumps(model)) / 1024


def measure_peak_memory(model, X_tr, y_tr, sw=None):
    """Peak RSS memory in MB consumed during model.fit() (tracemalloc)."""
    tracemalloc.start()
    if sw is not None:
        model.fit(X_tr, y_tr, sample_weight=sw)
    else:
        model.fit(X_tr, y_tr)
    _, peak = tracemalloc.get_traced_memory()
    tracemalloc.stop()
    return peak / 1024 / 1024


def fresh_xgb(seed):
    return XGBClassifier(random_state=seed, eval_metric='mlogloss',
                         tree_method='hist', n_jobs=-1, verbosity=0)

def fresh_rf(seed):
    return RandomForestClassifier(random_state=seed, n_estimators=100, n_jobs=-1)

def fresh_voting(seed):
    return VotingClassifier(
        estimators=[
            ('rf',  RandomForestClassifier(random_state=seed,
                                           n_estimators=100, n_jobs=-1)),
            ('xgb', XGBClassifier(random_state=seed, eval_metric='mlogloss',
                                  tree_method='hist', n_jobs=-1, verbosity=0)),
            ('gb',  GradientBoostingClassifier(random_state=seed))
        ], voting='soft'
    )

def fresh_rf_weighted(seed):
    return RandomForestClassifier(random_state=seed, n_estimators=100,
                                   class_weight='balanced', n_jobs=-1)


def fresh_voting_weighted(seed):
    return VotingClassifier(
        estimators=[
            ('rf',  RandomForestClassifier(random_state=seed, n_estimators=100,
                                           class_weight='balanced', n_jobs=-1)),
            ('xgb', XGBClassifier(random_state=seed, eval_metric='mlogloss',
                                  tree_method='hist', n_jobs=-1, verbosity=0)),
            ('gb',  GradientBoostingClassifier(random_state=seed))
        ], voting='soft'
    )


def fresh_model(name, seed, weighted=False):
    if 'Voting' in name:
        return fresh_voting_weighted(seed) if weighted else fresh_voting(seed)
    elif 'XGBoost' in name:
        return fresh_xgb(seed)   # XGBoost uses sample_weight at fit time
    else:
        return fresh_rf_weighted(seed) if weighted else fresh_rf(seed)


def gaussian_oversample(X, y, std=GAUSSIAN_SIGMA, seed=42):
    np.random.seed(seed)
    classes, counts = np.unique(y, return_counts=True)
    max_n = counts.max()
    Xl, yl = [X], [y]
    for cls, cnt in zip(classes, counts):
        if cnt < max_n:
            Xc  = X[y == cls]
            idx = np.random.choice(len(Xc), max_n - cnt, replace=True)
            Xl.append(Xc[idx] + np.random.normal(0, std, Xc[idx].shape))
            yl.append(np.full(max_n - cnt, cls))
    return np.vstack(Xl), np.concatenate(yl)


def apply_resampling(tech_name, X_tr, y_tr, seed, sigma=GAUSSIAN_SIGMA):
    if tech_name == 'No Resampling':
        return X_tr, y_tr, None, False
    elif tech_name == 'Gaussian Noise':
        X_res, y_res = gaussian_oversample(X_tr, y_tr, std=sigma, seed=seed)
        return X_res, y_res, None, False
    elif tech_name == 'Class Weighted':
        cls_counts  = np.bincount(y_tr)
        cls_weights = 1.0 / cls_counts
        sw = cls_weights[y_tr]
        sw = sw / sw.sum() * len(y_tr)
        return X_tr, y_tr, sw, True   # True = use class_weight="balanced" for RF/Voting
    elif tech_name == 'SMOTE':
        min_cnt   = np.bincount(y_tr).min()
        k         = min(5, min_cnt - 1)
        resampler = SMOTE(random_state=seed, k_neighbors=k)
        X_res, y_res = resampler.fit_resample(X_tr, y_tr)
        return X_res, y_res, None, False
    elif tech_name == 'ADASYN':
        min_cnt   = np.bincount(y_tr).min()
        n         = min(5, min_cnt - 1)
        resampler = ADASYN(random_state=seed, n_neighbors=n)
        X_res, y_res = resampler.fit_resample(X_tr, y_tr)
        return X_res, y_res, None, False
    elif tech_name == 'Random Undersample':
        resampler    = RandomUnderSampler(random_state=seed)
        X_res, y_res = resampler.fit_resample(X_tr, y_tr)
        return X_res, y_res, None, False
    else:
        raise ValueError(f"Unknown technique: {tech_name}")


def save_df(df, results_dir, filename):
    path = os.path.join(results_dir, f"{filename}.xlsx")
    df.to_excel(path, index=False)
    print(f"  Saved: {path}")


def save_fig(results_dir, filename):
    path = os.path.join(results_dir, f"{filename}.pdf")
    plt.savefig(path, dpi=300, bbox_inches='tight')
    print(f"  Saved: {path}")


# ============================================================
# DATA LOADING
# ============================================================

def load_data():
    print("=" * 80)
    print("LOADING IoT-RPL 2021 ENGINEERED DATASET")
    print("=" * 80)

    df_train = pd.read_csv(TRAIN_PATH, low_memory=False)
    df_test  = pd.read_csv(TEST_PATH,  low_memory=False)

    le = LabelEncoder()
    le.fit(pd.concat([df_train[LABEL_COL], df_test[LABEL_COL]]))

    X_train_full = df_train[FEATURE_COLS].values.astype(np.float32)
    y_train_full = le.transform(df_train[LABEL_COL])
    X_test_full  = df_test[FEATURE_COLS].values.astype(np.float32)
    y_test_full  = le.transform(df_test[LABEL_COL])

    X_test, y_test = stratified_sample(X_test_full, y_test_full,
                                        TEST_SAMPLE, seed=0)
    scaler = StandardScaler()
    scaler.fit(X_train_full)
    X_test_scaled = scaler.transform(X_test)

    print(f"  Full train   : {X_train_full.shape}")
    print(f"  Full test    : {X_test_full.shape}")
    print(f"  Test sample  : {X_test_scaled.shape}  (fixed, seed=0)")
    print(f"  Classes      : {list(le.classes_)}")
    print(f"  Seeds        : {SEEDS}")
    print(f"  Train/seed   : {TRAIN_SAMPLE:,}")
    print(f"\n  Natural class distribution (full train):")
    for cls, cnt in zip(*np.unique(y_train_full, return_counts=True)):
        lbl = le.inverse_transform([cls])[0]
        print(f"    {str(lbl).ljust(12)}: {cnt:,}  ({cnt/len(y_train_full)*100:.1f}%)")

    return X_train_full, y_train_full, X_test_scaled, y_test, scaler, le


# ============================================================
# BASELINE
# ============================================================

def run_baseline(X_train_full, y_train_full, X_test, y_test,
                 scaler, le, results_dir):
    print("\n" + "=" * 80)
    print("BASELINE")
    print("=" * 80)

    model_names  = ['Random Forest', 'XGBoost', 'Voting Classifier (Soft)']
    seed_results = {mn: [] for mn in model_names}

    for seed in SEEDS:
        print(f"\n  Seed {seed}:")
        X_tr_raw, y_tr = stratified_sample(X_train_full, y_train_full,
                                            TRAIN_SAMPLE, seed)
        X_tr = scaler.transform(X_tr_raw)

        for mname in model_names:
            model = fresh_model(mname, seed)
            t0    = time.perf_counter()
            model.fit(X_tr, y_tr)
            train_time = time.perf_counter() - t0
            peak_mb    = measure_peak_memory(fresh_model(mname, seed), X_tr, y_tr)

            y_pred      = model.predict(X_test)
            y_tr_pred   = model.predict(X_tr)
            acc         = accuracy_score(y_test, y_pred)
            train_acc   = accuracy_score(y_tr,   y_tr_pred)
            macro_f1    = f1_score(y_test, y_pred, average='macro',    zero_division=0)
            weighted_f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)
            gap         = train_acc - acc
            infer_us    = measure_inference_time(model, X_test)
            model_kb    = measure_model_size(model)

            seed_results[mname].append({
                'seed': seed, 'accuracy': acc, 'macro_f1': macro_f1,
                'weighted_f1': weighted_f1, 'overfitting_gap': gap,
                'train_time_s': train_time, 'infer_us': infer_us,
                'model_kb': model_kb, 'peak_mb': peak_mb,
            })
            print(f"    {mname.ljust(28)}: Acc={acc:.4f}, MacroF1={macro_f1:.4f}, "
                  f"Gap={gap:.4f}, Train={train_time:.1f}s, "
                  f"Infer={infer_us:.2f}µs, Size={model_kb:.1f}KB, "
                  f"Peak={peak_mb:.1f}MB")

    rows = []
    for mname in model_names:
        recs = seed_results[mname]
        rows.append({
            'Model'            : mname,
            'Accuracy'         : ci_str([r['accuracy']        for r in recs]),
            'Macro F1'         : ci_str([r['macro_f1']        for r in recs]),
            'Weighted F1'      : ci_str([r['weighted_f1']     for r in recs]),
            'Overfitting Gap'  : ci_str([r['overfitting_gap'] for r in recs]),
            'Train Time (s)'   : ci_str([r['train_time_s']    for r in recs]),
            'Infer (µs/sample)': ci_str([r['infer_us']        for r in recs]),
            'Model Size (KB)'  : ci_str([r['model_kb']        for r in recs]),
            'Peak Memory (MB)' : ci_str([r['peak_mb']         for r in recs]),
            '_acc_mean'        : np.mean([r['accuracy']        for r in recs]),
        })

    df = (pd.DataFrame(rows)
          .sort_values('_acc_mean', ascending=False)
          .drop(columns=['_acc_mean']))
    print("\n  BASELINE TABLE (mean [95% CI]):")
    print(df.to_string(index=False))
    save_df(df, results_dir, "Table1_Baseline_Results")

    rf_baseline_wf1 = np.mean([r['weighted_f1']
                                for r in seed_results['Random Forest']])
    return seed_results, rf_baseline_wf1


# ============================================================
# PHASE 1: IMBALANCE STUDY
# ============================================================

def run_imbalance_study(X_train_full, y_train_full, X_test, y_test,
                        scaler, le, results_dir):
    print("\n" + "=" * 80)
    print("PHASE 1: CLASS IMBALANCE STUDY")
    print("=" * 80)

    classes    = np.unique(y_train_full)
    normal_cls = le.transform(['Normal'])[0]
    attack_cls = [c for c in classes if c != normal_cls]
    class_names = list(le.classes_)

    scenarios  = ['Natural', 'Mild (70/30)', 'Moderate (85/15)', 'Severe (95/5)']
    techniques = ['No Resampling', 'SMOTE', 'ADASYN',
                  'Random Undersample', 'Gaussian Noise', 'Class Weighted']
    fractions  = {
        'Natural': None, 'Mild (70/30)': 0.70,
        'Moderate (85/15)': 0.85, 'Severe (95/5)': 0.95
    }

    seed_scores    = {sc: {t: [] for t in techniques} for sc in scenarios}
    per_class_last = {}

    for seed_idx, seed in enumerate(SEEDS):
        print(f"\n  Seed {seed} ({seed_idx+1}/{len(SEEDS)}):")
        np.random.seed(seed)

        for scenario in scenarios:
            if scenario == 'Natural':
                X_tr_raw, y_tr_base = stratified_sample(
                    X_train_full, y_train_full, TRAIN_SAMPLE, seed)
            else:
                nf      = fractions[scenario]
                n_norm  = int(TRAIN_SAMPLE * nf)
                n_atk   = int(TRAIN_SAMPLE * (1 - nf) / len(attack_cls))
                Xl, yl  = [], []
                for cls in classes:
                    Xc  = X_train_full[y_train_full == cls]
                    yc  = y_train_full[y_train_full == cls]
                    nr  = n_norm if cls == normal_cls else n_atk
                    idx = np.random.choice(len(Xc), nr, replace=(nr > len(Xc)))
                    Xl.append(Xc[idx]); yl.append(yc[idx])
                X_tr_raw  = np.vstack(Xl)
                y_tr_base = np.concatenate(yl)

            X_tr_base = scaler.transform(X_tr_raw)

            for tech in techniques:
                try:
                    X_res, y_res, sw, use_weighted = apply_resampling(
                        tech, X_tr_base, y_tr_base, seed)
                    model = fresh_xgb(seed)
                    if sw is not None:
                        model.fit(X_res, y_res, sample_weight=sw)
                    else:
                        model.fit(X_res, y_res)

                    y_pred   = model.predict(X_test)
                    macro_f1 = f1_score(y_test, y_pred,
                                        average='macro', zero_division=0)
                    seed_scores[scenario][tech].append(macro_f1)

                    if scenario == 'Severe (95/5)' and seed == SEEDS[-1]:
                        per_class_last[tech] = f1_score(
                            y_test, y_pred, average=None, zero_division=0)
                except Exception as e:
                    print(f"    [{scenario}] {tech}: FAILED — {e}")
                    seed_scores[scenario][tech].append(np.nan)

            line = f"    {scenario.ljust(18)}: "
            for tech in techniques:
                s = seed_scores[scenario][tech]
                if s:
                    line += f"{tech[:6]}={s[-1]:.3f}  "
            print(line)

    # Summary table
    rows = []
    for scenario in scenarios:
        for tech in techniques:
            scores = [s for s in seed_scores[scenario][tech] if not np.isnan(s)]
            if not scores:
                continue
            m, lo, hi = bootstrap_ci(scores)
            rows.append({
                'Scenario'        : scenario,
                'Technique'       : tech,
                'Macro F1 Mean'   : round(m,  4),
                'Macro F1 95% CI' : f"[{lo:.4f}, {hi:.4f}]",
                'Macro F1 Std'    : round(np.std(scores), 5),
            })

    df_summary = pd.DataFrame(rows)
    print("\n  PHASE 1 SUMMARY:")
    print(df_summary.to_string(index=False))
    save_df(df_summary, results_dir, "Table2_Phase1_Imbalance_Results")

    # Wilcoxon tests — Issue 4
    tech_pairs = [
        ('No Resampling', 'Class Weighted'),
        ('SMOTE',         'Class Weighted'),
        ('SMOTE',         'Gaussian Noise'),
        ('SMOTE',         'No Resampling'),
        ('ADASYN',        'Class Weighted'),
    ]
    wrows = []
    print("\n  WILCOXON TESTS (Severe scenario):")
    for t1, t2 in tech_pairs:
        s1  = seed_scores['Severe (95/5)'][t1]
        s2  = seed_scores['Severe (95/5)'][t2]
        p   = wilcoxon_test(s1, s2)
        sig = ('**' if (not np.isnan(p) and p < 0.01) else
               '*'  if (not np.isnan(p) and p < 0.05) else 'ns')
        print(f"    {t1.ljust(22)} vs {t2.ljust(22)}: p={p:.4f} {sig}")
        wrows.append({'Technique A': t1, 'Technique B': t2,
                      'Wilcoxon p':  round(p, 4) if not np.isnan(p) else 'n/a',
                      'Significant': sig})
    save_df(pd.DataFrame(wrows), results_dir, "Table2b_Wilcoxon_Tests")

    # Sigma sensitivity sweep — Issue 6
    print("\n  GAUSSIAN SIGMA SENSITIVITY SWEEP:")
    srows = []
    X_tr_raw_s, y_tr_s = stratified_sample(
        X_train_full, y_train_full, TRAIN_SAMPLE, SEEDS[0])
    X_tr_s = scaler.transform(X_tr_raw_s)
    nf_mild   = 0.70
    n_norm    = int(TRAIN_SAMPLE * nf_mild)
    n_atk     = int(TRAIN_SAMPLE * (1 - nf_mild) / len(attack_cls))
    Xl, yl    = [], []
    for cls in classes:
        Xc  = X_train_full[y_train_full == cls]
        yc  = y_train_full[y_train_full == cls]
        nr  = n_norm if cls == normal_cls else n_atk
        idx = np.random.choice(len(Xc), nr, replace=(nr > len(Xc)))
        Xl.append(Xc[idx]); yl.append(yc[idx])
    X_mild_raw = np.vstack(Xl)
    y_mild     = np.concatenate(yl)
    X_mild     = scaler.transform(X_mild_raw)

    for sigma in SIGMA_SWEEP:
        X_res, y_res = gaussian_oversample(X_mild, y_mild,
                                            std=sigma, seed=SEEDS[0])
        model = fresh_xgb(SEEDS[0])
        model.fit(X_res, y_res)
        mf1 = f1_score(model.predict(X_test), y_test,
                       average='macro', zero_division=0)
        print(f"    sigma={sigma}: MacroF1={mf1:.4f}")
        srows.append({'Sigma': sigma, 'Scenario': 'Mild (70/30)',
                      'Macro F1': round(mf1, 4)})
    save_df(pd.DataFrame(srows), results_dir, "Table2c_Sigma_Sensitivity")

    return seed_scores, per_class_last, class_names


# ============================================================
# PHASE 1 PLOTS
# ============================================================

def plot_phase1(seed_scores, per_class_last, class_names, results_dir):
    scenarios  = list(seed_scores.keys())
    techniques = list(seed_scores[scenarios[0]].keys())

    # Figure 1: Heatmap
    data = []
    for tech in techniques:
        row = []
        for sc in scenarios:
            scores = [s for s in seed_scores[sc][tech] if not np.isnan(s)]
            row.append(np.mean(scores) if scores else np.nan)
        data.append(row)

    df_h = pd.DataFrame(data, index=techniques, columns=scenarios)
    plt.figure(figsize=(13, 6))
    sns.heatmap(df_h, annot=True, fmt='.4f', cmap='RdYlGn',
                vmin=0.0, vmax=1.0, linewidths=0.5, annot_kws={"size": 10})
    plt.title('Figure 1: Mean Macro-F1 Heatmap\n'
              'Resampling Technique × Imbalance Scenario (XGBoost)')
    plt.xlabel('Imbalance Scenario')
    plt.ylabel('Resampling Technique')
    plt.tight_layout()
    save_fig(results_dir, "Figure1_Phase1_MacroF1_Heatmap")
    plt.show()

    # Figure 2: Line plot with CI
    fig, ax = plt.subplots(figsize=(12, 5))
    for tech in techniques:
        means, lo_errs, hi_errs = [], [], []
        for sc in scenarios:
            scores = [s for s in seed_scores[sc][tech] if not np.isnan(s)]
            if scores:
                m, lo, hi = bootstrap_ci(scores)
                means.append(m)
                lo_errs.append(m - lo)
                hi_errs.append(hi - m)
            else:
                means.append(np.nan)
                lo_errs.append(0); hi_errs.append(0)
        ax.errorbar(range(len(scenarios)), means,
                    yerr=[lo_errs, hi_errs],
                    marker='o', linewidth=2, capsize=4, label=tech)

    ax.set_xticks(range(len(scenarios)))
    ax.set_xticklabels(scenarios, rotation=10)
    ax.set_xlabel('Imbalance Scenario')
    ax.set_ylabel('Macro F1 (mean ± 95% CI)')
    ax.grid(True, alpha=0.3)
    ax.legend(fontsize=8)
    ax.set_title('Figure 2: Macro-F1 Under Increasing Normal-Dominance\n'
                 'Error bars = 95% bootstrap CI — XGBoost')
    plt.tight_layout()
    save_fig(results_dir, "Figure2_MacroF1_With_CI")
    plt.show()

    # Figure 3: Per-class F1 severe
    if per_class_last:
        x     = np.arange(len(class_names))
        width = 0.13
        plt.figure(figsize=(14, 5))
        for i, (tech, pc) in enumerate(per_class_last.items()):
            while len(pc) < len(class_names):
                pc = np.append(pc, 0)
            plt.bar(x + i * width, pc[:len(class_names)],
                    width, label=tech, alpha=0.85)
        plt.xlabel('Attack Class')
        plt.ylabel('F1 Score')
        plt.title('Figure 3: Per-Class F1 — Severe Imbalance (95% Normal) — XGBoost')
        plt.xticks(x + width * len(per_class_last) / 2,
                   class_names, rotation=10)
        plt.legend(loc='lower right', fontsize=8)
        plt.ylim(0, 1.15)
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        save_fig(results_dir, "Figure3_PerClass_F1_Severe")
        plt.show()

    # Figure 7: Per-seed scatter
    fig, axes = plt.subplots(1, len(scenarios), figsize=(16, 5), sharey=True)
    for ax, scenario in zip(axes, scenarios):
        for tech in techniques:
            scores = seed_scores[scenario][tech]
            ax.scatter(range(len(SEEDS)), scores,
                       label=tech, alpha=0.75, s=40)
        ax.set_title(scenario, fontsize=9)
        ax.set_xlabel('Seed index')
        ax.set_ylabel('Macro F1')
        ax.grid(True, alpha=0.3)
        ax.set_xticks(range(len(SEEDS)))
        ax.set_xticklabels([str(s) for s in SEEDS], rotation=45, fontsize=7)
    axes[0].legend(fontsize=7, loc='lower left')
    plt.suptitle('Figure 7: Per-Seed Macro F1 Scatter\n'
                 'Variability across seeds', fontsize=11)
    plt.tight_layout()
    save_fig(results_dir, "Figure7_Per_Seed_Scatter")
    plt.show()


# ============================================================
# PHASE 2: FEATURE RANKINGS
# ============================================================

def run_feature_rankings(X_train_full, y_train_full, X_test, y_test,
                          scaler, feature_names, results_dir):
    print("\n" + "=" * 80)
    print("PHASE 2: FEATURE RANKINGS (fixed subsample, seed=42)")
    print("=" * 80)

    X_s_raw, y_s = stratified_sample(X_train_full, y_train_full, 50_000, 42)
    X_s = scaler.transform(X_s_raw)

    # SHAP
    print("\n  SHAP ranking...")
    rf = fresh_rf(42)
    rf.fit(X_s, y_s)
    explainer = shap.TreeExplainer(rf)
    sv        = explainer.shap_values(X_test[:min(2000, len(X_test))])
    if isinstance(sv, list):
        mean_abs = np.mean([np.abs(s).mean(axis=0) for s in sv], axis=0)
    elif len(sv.shape) == 3:
        mean_abs = np.abs(sv).mean(axis=(0, 2))
    else:
        mean_abs = np.abs(sv).mean(axis=0)

    shap_df = (pd.DataFrame({'Feature': feature_names, 'SHAP': mean_abs})
               .sort_values('SHAP', ascending=False)
               .reset_index(drop=True))
    print(shap_df.to_string())
    shap_ranking = shap_df['Feature'].tolist()

    # RFE
    print("\n  RFE ranking...")
    rfe = RFE(RandomForestClassifier(n_estimators=50, random_state=42, n_jobs=-1),
              n_features_to_select=1, step=1)
    rfe.fit(X_s, y_s)
    rfe_df = (pd.DataFrame({'Feature': feature_names, 'RFE_Rank': rfe.ranking_})
              .sort_values('RFE_Rank')
              .reset_index(drop=True))
    print(rfe_df.to_string())
    rfe_ranking = rfe_df['Feature'].tolist()

    # Mutual Information
    print("\n  Mutual Information ranking...")
    mi    = mutual_info_classif(X_s, y_s, random_state=42)
    mi_df = (pd.DataFrame({'Feature': feature_names, 'MI': mi})
             .sort_values('MI', ascending=False)
             .reset_index(drop=True))
    print(mi_df.to_string())
    mi_ranking = mi_df['Feature'].tolist()

    # Ranking correlations — Issue 7
    print("\n  RANKING CORRELATIONS (Spearman rho + Kendall tau):")

    def to_pos(ranking):
        return [ranking.index(f) + 1 for f in feature_names]

    shap_pos = to_pos(shap_ranking)
    rfe_pos  = to_pos(rfe_ranking)
    mi_pos   = to_pos(mi_ranking)

    pairs = [('SHAP', 'RFE',        shap_pos, rfe_pos),
             ('SHAP', 'MutualInfo', shap_pos, mi_pos),
             ('RFE',  'MutualInfo', rfe_pos,  mi_pos)]
    corr_rows = []
    print(f"\n  {'Pair'.ljust(28)} Spearman rho    Kendall tau")
    for m1, m2, p1, p2 in pairs:
        rho, rho_p = spearmanr(p1, p2)
        tau, tau_p = kendalltau(p1, p2)
        label  = f"{m1} vs {m2}"
        interp = ('Strong'   if abs(rho) > 0.8 else
                  'Moderate' if abs(rho) > 0.5 else 'Weak')
        print(f"  {label.ljust(28)} rho={rho:.4f}(p={rho_p:.3f})  "
              f"tau={tau:.4f}(p={tau_p:.3f})  [{interp}]")
        corr_rows.append({'Method Pair': label,
                          'Spearman rho': round(rho,   4),
                          'Spearman p':   round(rho_p, 4),
                          'Kendall tau':  round(tau,   4),
                          'Kendall p':    round(tau_p, 4),
                          'Interpretation': interp})
    save_df(pd.DataFrame(corr_rows), results_dir, "Table5_Ranking_Correlations")

    # Figure 4
    fig, axes = plt.subplots(1, 3, figsize=(16, 5))
    axes[0].barh(shap_df['Feature'][::-1], shap_df['SHAP'][::-1],
                 color='steelblue', alpha=0.85)
    axes[0].set_title('SHAP')
    axes[0].set_xlabel('Mean |SHAP Value|')
    axes[0].grid(True, alpha=0.3)

    axes[1].barh(rfe_df['Feature'][::-1], rfe_df['RFE_Rank'][::-1],
                 color='seagreen', alpha=0.85)
    axes[1].set_title('RFE (lower = more important)')
    axes[1].set_xlabel('Rank')
    axes[1].grid(True, alpha=0.3)

    axes[2].barh(mi_df['Feature'][::-1], mi_df['MI'][::-1],
                 color='darkorange', alpha=0.85)
    axes[2].set_title('Mutual Information')
    axes[2].set_xlabel('MI Score')
    axes[2].grid(True, alpha=0.3)

    plt.suptitle('Figure 4: Feature Ranking — SHAP, RFE, Mutual Information',
                 fontsize=12)
    plt.tight_layout()
    save_fig(results_dir, "Figure4_Feature_Ranking_Comparison")
    plt.show()

    return ({'SHAP': shap_ranking, 'RFE': rfe_ranking, 'MutualInfo': mi_ranking},
            shap_df, rfe_df, mi_df)


# ============================================================
# PHASE 2: SUBSET EVALUATION + EFFICIENCY (Issue 3)
# ============================================================

def run_subset_evaluation(rankings, X_train_full, y_train_full,
                           X_test, y_test, scaler, feature_names,
                           rf_baseline_wf1, le, results_dir):
    print("\n" + "=" * 80)
    print("PHASE 2: FEATURE SUBSET EVALUATION + EFFICIENCY METRICS")
    print("=" * 80)

    model_names = ['Random Forest', 'XGBoost', 'Voting Classifier (Soft)']

    seed_subset = {
        m: {n: {mn: [] for mn in model_names}
            for n in SUBSET_SIZES}
        for m in rankings
    }

    for seed in SEEDS:
        print(f"\n  Seed {seed}:")
        X_tr_raw, y_tr = stratified_sample(X_train_full, y_train_full,
                                            TRAIN_SAMPLE, seed)
        X_tr = scaler.transform(X_tr_raw)

        for method, ranking in rankings.items():
            for n in SUBSET_SIZES:
                feat_idx = [feature_names.index(f) for f in ranking[:n]]
                X_tr_sub = X_tr[:, feat_idx]
                X_te_sub = X_test[:, feat_idx]

                for mname in model_names:
                    try:
                        model = fresh_model(mname, seed)
                        t0    = time.perf_counter()
                        model.fit(X_tr_sub, y_tr)
                        train_s  = time.perf_counter() - t0
                        peak_mb  = measure_peak_memory(
                            fresh_model(mname, seed), X_tr_sub, y_tr)

                        y_pred   = model.predict(X_te_sub)
                        wf1      = f1_score(y_test, y_pred,
                                            average='weighted', zero_division=0)
                        infer_us = measure_inference_time(model, X_te_sub)
                        model_kb = measure_model_size(model)

                        seed_subset[method][n][mname].append({
                            'wf1': wf1, 'train_s': train_s,
                            'infer_us': infer_us, 'model_kb': model_kb,
                            'peak_mb': peak_mb,
                        })
                    except Exception as e:
                        print(f"    [{method}] n={n} {mname}: FAILED — {e}")

        for method in rankings:
            line = f"    {method}: "
            for n in SUBSET_SIZES:
                recs = seed_subset[method][n]['XGBoost']
                if recs:
                    line += f"n={n}→{recs[-1]['wf1']:.3f}  "
            print(line)

    # Build tables
    rows     = []
    eff_rows = []
    n_total  = len(feature_names)

    for method in rankings:
        for n in SUBSET_SIZES:
            for mname in model_names:
                recs = seed_subset[method][n][mname]
                if not recs:
                    continue
                overhead = int((n_total - n) / n_total * 100)
                wf1s     = [r['wf1']      for r in recs]
                train_ss = [r['train_s']  for r in recs]
                infer_us = [r['infer_us'] for r in recs]
                kbs      = [r['model_kb'] for r in recs]
                m_wf1, lo_wf1, hi_wf1 = bootstrap_ci(wf1s)

                rows.append({
                    'Selection Method'       : method,
                    'N Features'             : n,
                    'Model'                  : mname,
                    'Weighted F1'            : f"{m_wf1:.4f} [{lo_wf1:.4f}, {hi_wf1:.4f}]",
                    'Weighted F1 Std'        : round(np.std(wf1s), 5),
                    'Overhead Reduction (%)'   : f"{overhead}%",
                })
                eff_rows.append({
                    'Selection Method'       : method,
                    'N Features'             : n,
                    'Model'                  : mname,
                    'Train Time (s)'         : ci_str(train_ss),
                    'Infer Time (µs/sample)' : ci_str(infer_us),
                    'Model Size (KB)'        : ci_str(kbs),
                    'Peak Memory (MB)'       : ci_str([r['peak_mb'] for r in recs]),
                    'Overhead Reduction (%)'   : f"{overhead}%",
                })

    df_subset = pd.DataFrame(rows)
    df_eff    = pd.DataFrame(eff_rows)

    print("\n  PHASE 2 TABLE — SHAP:")
    shap_view = df_subset[df_subset['Selection Method'] == 'SHAP'].drop(
        columns=['Selection Method']).reset_index(drop=True)
    print(shap_view.to_string(index=False))

    save_df(df_subset, results_dir, "Table3_Phase2_Feature_Subset_Results")
    save_df(df_eff,    results_dir, "Table3b_Efficiency_Metrics")

    # Per-stratum — Issue 8
    _run_per_stratum(rankings, X_train_full, y_train_full,
                     X_test, y_test, scaler, feature_names, le, results_dir)

    # Figure 5
    methods = list(rankings.keys())
    fig, axes = plt.subplots(1, len(methods), figsize=(16, 5), sharey=True)
    for ax, method in zip(axes, methods):
        for mname in model_names:
            means, lo_errs, hi_errs = [], [], []
            for n in SUBSET_SIZES:
                recs = seed_subset[method][n][mname]
                if recs:
                    wf1s = [r['wf1'] for r in recs]
                    m, lo, hi = bootstrap_ci(wf1s)
                    means.append(m)
                    lo_errs.append(m - lo)
                    hi_errs.append(hi - m)
                else:
                    means.append(np.nan)
                    lo_errs.append(0); hi_errs.append(0)
            ax.errorbar(SUBSET_SIZES, means,
                        yerr=[lo_errs, hi_errs],
                        marker='o', linewidth=2, capsize=4, label=mname)
        ax.axhline(y=rf_baseline_wf1, color='gray', linestyle='--', alpha=0.6,
                   label=f'RF baseline ({rf_baseline_wf1:.4f})')
        ax.set_title(method)
        ax.set_xlabel('Number of Features')
        ax.set_ylabel('Weighted F1 (mean ± 95% CI)')
        ax.set_xticks(SUBSET_SIZES)
        ax.grid(True, alpha=0.3)
        ax.legend(fontsize=8)
    plt.suptitle(f'Figure 5: Weighted F1 vs Feature Subset Size\n'
                 f'Dashed = RF baseline ({rf_baseline_wf1:.4f})', fontsize=12)
    plt.tight_layout()
    save_fig(results_dir, "Figure5_WeightedF1_vs_Features")
    plt.show()

    # Figure 8: Inference time vs subset — Issue 3
    fig, axes = plt.subplots(1, len(methods), figsize=(16, 5), sharey=True)
    for ax, method in zip(axes, methods):
        for mname in model_names:
            means, lo_errs, hi_errs = [], [], []
            for n in SUBSET_SIZES:
                recs = seed_subset[method][n][mname]
                if recs:
                    vals = [r['infer_us'] for r in recs]
                    m, lo, hi = bootstrap_ci(vals)
                    means.append(m)
                    lo_errs.append(m - lo)
                    hi_errs.append(hi - m)
                else:
                    means.append(np.nan)
                    lo_errs.append(0); hi_errs.append(0)
            ax.errorbar(SUBSET_SIZES, means,
                        yerr=[lo_errs, hi_errs],
                        marker='s', linewidth=2, capsize=4, label=mname)
        ax.set_title(method)
        ax.set_xlabel('Number of Features')
        ax.set_ylabel('Inference Time (µs/sample)')
        ax.set_xticks(SUBSET_SIZES)
        ax.grid(True, alpha=0.3)
        ax.legend(fontsize=8)
    plt.suptitle('Figure 8: Inference Time vs Feature Subset Size\n'
                 'Error bars = 95% bootstrap CI', fontsize=12)
    plt.tight_layout()
    save_fig(results_dir, "Figure8_Inference_Time_vs_Features")
    plt.show()

    return seed_subset


# ============================================================
# ISSUE 8: PER-STRATUM BY ATTACK CLASS
# ============================================================

def _run_per_stratum(rankings, X_train_full, y_train_full,
                     X_test, y_test, scaler, feature_names, le, results_dir):
    print("\n  PER-STRATUM RESULTS (Issue 8) — full vs top-7 SHAP:")

    ranking       = rankings['SHAP']
    n_full        = len(feature_names)
    n_red         = 7
    feat_idx_full = list(range(n_full))
    feat_idx_red  = [feature_names.index(f) for f in ranking[:n_red]]

    X_tr_raw, y_tr = stratified_sample(X_train_full, y_train_full,
                                        TRAIN_SAMPLE, SEEDS[0])
    X_tr = scaler.transform(X_tr_raw)

    rows = []
    for subset_name, feat_idx in [('Full (13)', feat_idx_full),
                                   (f'Top-{n_red} SHAP', feat_idx_red)]:
        model = fresh_xgb(SEEDS[0])
        model.fit(X_tr[:, feat_idx], y_tr)
        y_pred      = model.predict(X_test[:, feat_idx])
        per_class   = f1_score(y_test, y_pred, average=None, zero_division=0)
        class_names = list(le.classes_)

        print(f"\n    {subset_name}:")
        for cls_name, f1_val in zip(class_names, per_class):
            print(f"      {cls_name.ljust(12)}: F1={f1_val:.4f}")
            rows.append({'Feature Subset': subset_name,
                         'Attack Class':   cls_name,
                         'F1 Score':       round(f1_val, 4)})

    df_strat = pd.DataFrame(rows)
    save_df(df_strat, results_dir, "Table3c_Per_Stratum_Results")

    pivot = df_strat.pivot(index='Attack Class',
                            columns='Feature Subset',
                            values='F1 Score')
    pivot.plot(kind='bar', figsize=(10, 5), rot=0, alpha=0.85)
    plt.title('Per-Stratum F1: Full vs Top-7 SHAP Features\n'
              'XGBoost, Natural distribution')
    plt.ylabel('F1 Score')
    plt.xlabel('Attack Class')
    plt.legend(title='Feature Subset')
    plt.grid(True, alpha=0.3, axis='y')
    plt.tight_layout()
    save_fig(results_dir, "Figure_Per_Stratum")
    plt.show()


# ============================================================
# COMBINED FIGURE
# ============================================================

def run_combined_figure(rankings, X_train_full, y_train_full,
                         X_test, y_test, scaler, feature_names,
                         le, results_dir):
    print("\n" + "=" * 80)
    print("PHASE 1+2 COMBINED: KEY FIGURE")
    print("=" * 80)

    scenarios  = ['Natural', 'Mild (70/30)', 'Moderate (85/15)', 'Severe (95/5)']
    normal_cls = le.transform(['Normal'])[0]
    attack_cls = [c for c in np.unique(y_train_full) if c != normal_cls]
    ranking    = rankings['SHAP']
    fractions  = {'Natural': None, 'Mild (70/30)': 0.70,
                  'Moderate (85/15)': 0.85, 'Severe (95/5)': 0.95}

    data = [[[] for _ in scenarios] for _ in SUBSET_SIZES]

    for seed in SEEDS:
        np.random.seed(seed)
        print(f"  Seed {seed}:")

        for j, scenario in enumerate(scenarios):
            if scenario == 'Natural':
                X_tr_raw, y_tr = stratified_sample(
                    X_train_full, y_train_full, TRAIN_SAMPLE, seed)
            else:
                nf     = fractions[scenario]
                n_norm = int(TRAIN_SAMPLE * nf)
                n_atk  = int(TRAIN_SAMPLE * (1 - nf) / len(attack_cls))
                Xl, yl = [], []
                for cls in np.unique(y_train_full):
                    Xc  = X_train_full[y_train_full == cls]
                    yc  = y_train_full[y_train_full == cls]
                    nr  = n_norm if cls == normal_cls else n_atk
                    idx = np.random.choice(len(Xc), nr, replace=(nr > len(Xc)))
                    Xl.append(Xc[idx]); yl.append(yc[idx])
                X_tr_raw = np.vstack(Xl)
                y_tr     = np.concatenate(yl)

            X_tr_scaled = scaler.transform(X_tr_raw)

            for i, n in enumerate(SUBSET_SIZES):
                feat_idx = [feature_names.index(f) for f in ranking[:n]]
                X_tr_sub = X_tr_scaled[:, feat_idx]
                X_te_sub = X_test[:, feat_idx]

                if scenario != 'Natural':
                    X_tr_sub, y_tr_sub = gaussian_oversample(
                        X_tr_sub, y_tr, seed=seed)
                else:
                    y_tr_sub = y_tr

                model = fresh_xgb(seed)
                model.fit(X_tr_sub, y_tr_sub)
                f1 = f1_score(model.predict(X_te_sub), y_test,
                              average='macro', zero_division=0)
                data[i][j].append(f1)

        n_means = [f"{np.mean(data[0][j]):.3f}" for j in range(len(scenarios))]
        print(f"    Top-3: {n_means}")

    mean_data = np.array([[np.mean(data[i][j])
                           for j in range(len(scenarios))]
                          for i in range(len(SUBSET_SIZES))])

    df_heatmap = pd.DataFrame(
        mean_data,
        index  =[f'Top {n} features' for n in SUBSET_SIZES],
        columns=scenarios
    )
    print("\n  COMBINED HEATMAP:")
    print(df_heatmap.round(4).to_string())
    save_df(df_heatmap.reset_index().rename(columns={'index': 'Feature Subset'}),
            results_dir, "Table4_Combined_Heatmap")

    plt.figure(figsize=(12, 5))
    sns.heatmap(df_heatmap, annot=True, fmt='.4f', cmap='RdYlGn',
                vmin=0.0, vmax=1.0, linewidths=0.5, annot_kws={"size": 12})
    plt.title('Figure 6: Mean Macro F1 — Feature Subset × Imbalance Scenario\n'
              'XGBoost | SHAP | Gaussian Noise (constructed) / Stratified (natural)',
              fontsize=10)
    plt.xlabel('Imbalance Scenario')
    plt.ylabel('Feature Subset')
    plt.tight_layout()
    save_fig(results_dir, "Figure6_Combined_Key_Figure")
    plt.show()


# ============================================================
# MAIN
# ============================================================

def main():
    print("=" * 80)
    print("RPL ATTACK DETECTION — IoT-RPL 2021 (COMPLETE FINAL)")
    print(f"Seeds: {SEEDS}  |  Train/seed: {TRAIN_SAMPLE:,}  |  Test: {TEST_SAMPLE:,}")
    print("=" * 80)

    timestamp   = datetime.now().strftime("%Y%m%d_%H%M%S")
    results_dir = f"RPL_Results_{timestamp}"
    os.makedirs(results_dir, exist_ok=True)
    print(f"Results directory: {results_dir}")

    X_train_full, y_train_full, X_test, y_test, scaler, le = load_data()

    _, rf_baseline_wf1 = run_baseline(
        X_train_full, y_train_full, X_test, y_test, scaler, le, results_dir)

    seed_scores_imb, per_class_last, class_names = run_imbalance_study(
        X_train_full, y_train_full, X_test, y_test, scaler, le, results_dir)

    plot_phase1(seed_scores_imb, per_class_last, class_names, results_dir)

    rankings, *_ = run_feature_rankings(
        X_train_full, y_train_full, X_test, y_test,
        scaler, FEATURE_COLS, results_dir)

    run_subset_evaluation(
        rankings, X_train_full, y_train_full, X_test, y_test,
        scaler, FEATURE_COLS, rf_baseline_wf1, le, results_dir)

    run_combined_figure(
        rankings, X_train_full, y_train_full, X_test, y_test,
        scaler, FEATURE_COLS, le, results_dir)

    print("\n" + "=" * 80)
    print(f"ALL DONE — results in: {results_dir}")
    print("=" * 80)
    print("\nOutputs:")
    for f in [
        "Table1_Baseline_Results.xlsx",
        "Table2_Phase1_Imbalance_Results.xlsx",
        "Table2b_Wilcoxon_Tests.xlsx",
        "Table2c_Sigma_Sensitivity.xlsx",
        "Table3_Phase2_Feature_Subset_Results.xlsx",
        "Table3b_Efficiency_Metrics.xlsx",
        "Table3c_Per_Stratum_Results.xlsx",
        "Table4_Combined_Heatmap.xlsx",
        "Table5_Ranking_Correlations.xlsx",
        "Figure1_Phase1_MacroF1_Heatmap.pdf",
        "Figure2_MacroF1_With_CI.pdf",
        "Figure3_PerClass_F1_Severe.pdf",
        "Figure4_Feature_Ranking_Comparison.pdf",
        "Figure5_WeightedF1_vs_Features.pdf",
        "Figure6_Combined_Key_Figure.pdf",
        "Figure7_Per_Seed_Scatter.pdf",
        "Figure8_Inference_Time_vs_Features.pdf",
        "Figure_Per_Stratum.pdf",
    ]:
        print(f"  {f}")


if __name__ == "__main__":
    main()
```
